```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LinearRegression

data = pd.read_csv('Employees_Details.csv')
#data.head(5)
data.describe()
plt.scatter( data['Experience'], data['Salary kUSD'], color='red')
plt.title('Experience vs salary', fontsize=14)
plt.xlabel('Experience', fontsize=14)
plt.ylabel('Salary kUSD', fontsize=14)
plt.grid(True)
#plt.show()
#lr  = LinearRegression().fit(data['Age'], data['Experience'])
#lr.score(data['Age'], data['Experience'])

data.corr()
df = pd.DataFrame(data, columns=['Employee Name', 'Employee ID', 'Age', 'Department','Location',
                                 'Experience','ML','Python', 'Salary kUSD'])
X=df[['Experience']]
Y=df['Salary kUSD']
regr = LinearRegression()
regr.fit(X,Y)
print('Intercept:  ', regr.intercept_)
print('coefficienets : ', regr.coef_)
print('Predicted salary :', regr.predict([[4]]))

```
